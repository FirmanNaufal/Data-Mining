```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
```

```python
dataset = 'D:\Kuliah\SEM 5\Data Mining\Tugas pertemuan 7\kindey stone urine analysis.csv'
```

```python
df = pd.read_csv(dataset)
df.head()
```

```python
feature = df.iloc[:, 0:-1]
target = df.iloc[:, -1]
# print(target)
print(feature)
```

```python
decisiontree = DecisionTreeClassifier(random_state=0, max_depth=None, min_samples_split=2, min_samples_leaf=1, min_weight_fraction_leaf=0, max_leaf_nodes=None, min_impurity_decrease=0)
```

```python
model = decisiontree.fit(feature,target)
```

```python
observation = pd.DataFrame([[1, 2, 3, 2, 5, 4]], columns=feature.columns)
prediction = model.predict(observation)
prediction_proba = model.predict_proba(observation)
```

```python
print("Prediksi:", prediction)
print("Probabilitas prediksi:", prediction_proba)
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree
plt.figure(figsize=(20, 10))
plot_tree(decisiontree, feature_names=feature.columns, class_names=['no', 'yes'], filled=True)
plt.savefig('decision-tree.png', format='png')
plt.show()
```
